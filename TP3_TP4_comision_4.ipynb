* HORAS DE EJERCICIOS diario= un número decimal

REalizar las siguientes indicaciones y extraer concluisiones sobre la información y la pooblación muestreada.

1.   Generar un array de dimensión 2 (matriz), con todos los datos ordenados, permitiendo visualizarlos como tabla.
2.   Suponiendo que la distribución de probabilidad de los datos muestrados de las N personas, se distribuyen en forma NORMAL. Encontrar la media en los datos de altura, peso, y horas de descanso, trabajo y ejercicios.
3.  Graficar la muestra o distribución de los datos  de pesos , y alturas en graficas separadas. Indicar el grado de variabilidad de sus datos
4.  Plotear de manera conjunta y encontrar si presentan correlación los datos de peso vs altura,
5. Plotear de manera conjunta los datos de horas de descanso, estudio y ejercicio



```python
import numpy as np
import matplotlib as plt

def generar_datos(N):
  sexo=np.random.choice(["Masculino","Femenino","No Especifica"],N,[0.4,0.5,0.1])

  edad=np.random.normal(44,6,N)
  edad=np.clip(np.around(edad),28,60)

  altura=np.random.normal(180,10,N)
  altura=np.clip(np.around(altura,decimals=1),160.0,200.0)

  peso=np.random.normal(80,15,N)
  peso=np.clip(np.around(peso),50.0,110.0)

  horas_trabajo=np.random.normal(10,2,N)
  horas_trabajo=np.clip(np.around(horas_trabajo,decimals=1),4.0,12.0)

  horas_ejercicio=np.random.normal(1,0.5,N)
  horas_ejercicio=np.clip(np.around(horas_ejercicio,decimals=1),0,2)

  horas_descanso= 24- horas_trabajo - horas_ejercicio

  datos = np.column_stack((sexo , edad, altura, peso, horas_trabajo, horas_ejercicio, horas_descanso))
  return datos

N=1000
datos=generar_datos(N)
```

```python
#detectar los valores atipicos
def detect_outliers_iqr(data, column):
    q1 = data[column].quantile(0.25)
    q3 = data[column].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers

# Llamamos a la función para cada columna numérica donde desees detectar outliers
columns_to_check = ["sepal.width"]
outliers_list = []
for col in columns_to_check:
    outliers = detect_outliers_iqr(df, col)
    outliers_list.append(outliers)

# Concatenamos todos los outliers encontrados en un solo DataFrame
outliers_df = pd.concat(outliers_list)


# Eliminamos los outliers del conjunto de datos original
#Drop: Elimina las filas segun los indices enviados por parametros
data_cleaned = df.drop(outliers_df.index)

# Imprimimos la cantidad de outliers detectados
print("Número de outliers detectados:", len(outliers_df))
```

```python
from matplotlib.colors import ListedColormap
import numpy as np

# Define las distintas variedades presentes en la columna "variety"
varieties = df['variety'].unique()

# Crea un mapeo de colores con una paleta de colores para el número de variedades
cmap = plt.get_cmap('tab10', len(varieties))

# Asigna un color único a cada variedad utilizando el mapeo de colores
colors = [cmap(i) for i in range(len(varieties))]
df['color'] = df['variety'].map(dict(zip(varieties, colors)))


colors2= [cmap(i) for i in range(len(varieties))]
data_cleaned['color'] = data_cleaned['variety'].map(dict(zip(varieties, colors2)))
# Crea el gráfico de dispersión usando los colores asignados
plt.scatter(df['sepal.length'], df['sepal.width'], c=df['color'])

# Etiqueta los ejes y muestra el gráfico
plt.xlabel('Longitud del sépalo')
plt.ylabel('Anchura del sépalo')
plt.ylim(1.5,4.5)
plt.title('Gráfico de dispersión de longitud y anchura del sépalo por variedad')
plt.show()
```

```python
# Análisis de frecuencias
df['variety'].value_counts().plot(kind='bar')
plt.xlabel('variety')
plt.ylabel('Frequency')
plt.title('Frequency Analysis: Species')
plt.show()
```

```python
#Diagrama de torta
labels = df['variety'].value_counts().index
counts = df['variety'].value_counts().values

# Crear el gráfico de torta
plt.figure(figsize=(8, 8))  # Puedes ajustar el tamaño del gráfico según tus preferencias

# Puedes cambiar los colores de las porciones si lo deseas

plt.pie(counts, labels=labels, autopct='%1.1f%%', startangle=90)
plt.axis('equal')  # Esto asegura que el gráfico de torta sea un círculo en lugar de una elipse
plt.title('Frequency Analysis: Species')
plt.show()
```

```python
fig, axs = plt.subplots(4, 2, figsize=(12, 10))

# Trazar el histograma y el boxplot para "Sepal Length"
axs[0, 0].hist(data_cleaned['sepal.length'])
axs[0, 0].set_xlabel('Sepal Length')
axs[0, 0].set_title('Histogram: Sepal Length')

axs[0, 1].boxplot(data_cleaned['sepal.length'],vert=False)
axs[0, 1].set_xlabel('Sepal Length')
axs[0, 1].set_title('Boxplot: Sepal Length')

# Trazar el histograma y el boxplot para "Sepal Width"
axs[1, 0].hist(data_cleaned['sepal.width'])
axs[1, 0].set_xlabel('Sepal Width')
axs[1, 0].set_title('Histogram: Sepal Width')

axs[1, 1].boxplot(data_cleaned['sepal.width'],vert=False)
axs[1, 1].set_xlabel('Sepal Width')
axs[1, 1].set_title('Boxplot: Sepal Width')


# Trazar el histograma y el boxplot para "Petal Length"
axs[2, 0].hist(data_cleaned['petal.length'])
axs[2, 0].set_xlabel('Petal Length')
axs[2, 0].set_title('Histogram: Petal Length')

axs[2, 1].boxplot(data_cleaned['petal.length'],vert=False)
axs[2, 1].set_xlabel('Petal Length')
axs[2, 1].set_title('Boxplot: Petal Length')

# Trazar el histograma y el boxplot para "Sepal Width"
axs[3, 0].hist(data_cleaned['petal.width'])
axs[3, 0].set_xlabel('petal Width')
axs[3, 0].set_title('Histogram: petal Width')

axs[3, 1].boxplot(data_cleaned['petal.width'],vert=False)
axs[3, 1].set_xlabel('petal Width')
axs[3, 1].set_title('Boxplot: petal Width')

plt.tight_layout()

plt.show()
```

# Ejercicio 8

Realizar analisis de matriz corelacion y explicar que variable estan correlacionadas

```python
correlation_matrix = df_clear[['sepal.length', 'sepal.width', 'petal.length', 'petal.width']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
